## KDD 2019

```python
data = pd.read_csv('input/kdd2019_state.csv')
```

```python
data.head()
```

```python
data.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
data.shape
```

```python
#categorical_features = ['col3', 'col4']
#numeric_features = ['col1', 'col2']
```

```python
ctgan = CTGANSynthesizer(verbose=True)
#ctgan.fit(data, categorical_features, epochs = 300)
ctgan.fit(data, epochs = 300)
```

```python
samples = ctgan.sample(3000)
print(samples.head())
```

```python
samples.shape
```

```python
print(data.shape, samples.shape)
```

```python
samples.to_csv('synthetic_kdd.csv')
```

```python
table_evaluator =  TableEvaluator(data, samples)
```

```python
table_evaluator.visual_evaluation()
```
