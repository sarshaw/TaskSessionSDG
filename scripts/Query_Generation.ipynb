```python
# load the libraries

import os
import pandas as pd
import numpy as np

!pip install keybert
from keybert import KeyBERT
```

```python
# load the input file - list of topics

data = pd.read_csv("./data/session2014_topic.csv", encoding = 'latin')
```

```python
# first query segment generation

data['query'] = 0

kw_model = KeyBERT()

for i in data.index:
     text = data['session_topic'][i]
     keywords = kw_model.extract_keywords(text)   
     data['query'][i] = keywords
```

```python
# create session and query segment ids

data['query_id'] = 0
data['session_id'] = range(0, len(data))
```
